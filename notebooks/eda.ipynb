```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import avalanche
import seaborn as sns
```

# Value Count Plots:

```python
dftrain=pd.read_csv('../data/train_MachineLearningCVE.csv')
df5_train=pd.read_csv('../data/fiveclass_train.csv')
df2_train=pd.read_csv('../data/bin_train.csv')
```

```python

plt.figure(figsize=(8,8))
ax=df5_train.fiveLabels.value_counts().plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.title("Count of different type of attacks after grouping into five")
plt.xlabel('Attack Types')
plt.ylabel('Attack Count')
```
